```python
def hypothesis(theta,x):
  return theta[0]+theta[1]*x



def gradients(X,Y,theta):

  m=X.shape[0];

  grad=np.zeros((2,))

  for i in range(m):

    y_=hypothesis(theta,X[i])

    y=Y[i]

    grad[0]=grad[0]+(y_-y)

    grad[1]=grad[1]+(y_-y)*X[i]

  return grad/m



def error(X,Y,theta):

  m=X.shape[0]

  total_error=0.0

  for i in range(m):

    y_=hypothesis(theta,X[i])

    total_error+=(y_-Y[i])**2

  return total_error/m




def gradientDescent(X,Y,learning_rate=0.1):

  theta=np.zeros((2,))

  error_list=[]


  for i in range(100):

    grad=gradients(X,Y,theta)

    e=error(X,Y,theta)

    error_list.append(e)

    theta[0]=theta[0]-learning_rate*grad[0]

    theta[1]=theta[1]-learning_rate*grad[1]




  return theta,error_list


```

```python
theta,error_list=gradientDescent(X,Y)
```

```python
plt.scatter()
plt.plot(error_list)
```
